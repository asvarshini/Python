```python
#3(b) multilevel inheritance
class grandparents:
    def __init__(self, colour):
        self.colour = colour

    def re(self):
        return self.colour


class parent(grandparents):
    def __init__(self, colour, hair):
        super().__init__(colour)
        self.hair = hair

    def rea(self):
        return self.hair


class child(parent):
    def __init__(self, study, colour, hair):
        super().__init__(colour, hair)
        self.study = study

    def read(self):
        return self.re(), self.rea(), self.study


s3 = child("good", "black", "long")
print(s3.read())
```

```python
#4(a)Hierachical inheritance
class par:
  def __init__(self,gold,pro):
    self.gold =gold
    self.pro =pro
class sis1(par):
  def __init__(self,gold,pro,n):
    par.__init__(self,gold,pro)
    self.n =n
    print(gold,pro,n)
class sis2(par):
  def __init__(self,gold,pro,n):
    par.__init__(self,gold,pro)
    self.n =n
    print(gold,pro,n)
s1=sis1("24k",10,5)
s2=sis2("18k",12,22)



```
